__Confusion Matrix__:
Lets define one new metric for the estimation of predictive power of the model called accuracy. It determines how much percentage of the predictions have been made correct using the model meaning predictions meeting the actual classified labels. In the process we create an confusion matrix. The logic behind in creating the confusion matrix is there might be possibility of miss classification happening in predictions. Also the result of the logistic function is in terms of probabilities so always there might be chance of making type 1 and type 2 errors. These forms the basis of creation of confusion matrix. In the matrix we have actual labels in the horizontal and predicted label in the vertical and there exists some combinations in the matrix in the form 2x2. The matrix clearly determines how many true actual values have been predicted true and false correspondingly. Similarly it also determines how many false actual values have been predicted false and true correspondingly. This confusion amtrix visualization is just for the undertsanding poiny of view. In real python coding we don't get any sort of visual understanding.

Apart from accuracy there is neccessity of estimating other metrics in order to assess the model as using accuracy alone we cannot completely assess the model as accuracy accounts for both postives and neagtive effect of classes collectively. But sometimes as per business requirment we need to concentrate on only one class variable. In order to do so we calculate other metrics using confusion matrix like __Sensitivity__ and __Specificity__. __Sensitivity__ determines the ratio of number of actual yess being correctly predicted by the model to the number of total actual yess. Through this metric we can get to True positive rate of the model. Using this we can check how the yess or positives are being predicted by the model. __Specificity__ accounts for ratio of number of actual nos being correctly predicted by the model to the total number of actual nos. Through this we can get the value of (1- FPR). FPR stands for __False postitive rate__, with which we can get to know how many positives have been predicted from actual negatives. Apart from these metrics there are other useful metrics like __True predictive power(Precision)__ and __False predictive power__ of the model.

Specificity is considerably high about 90%, but our requirment in this telecom problem is not identifying not churn customers hence we dont take much  into consideration of specificity value as per this problem description. Hence there is requirement of selected best threshold probability value such that sensitivity is as high as possible and False positive rate is as low as possible.

#### Estimating optimum threshold probability value

```python
# creating a range of threshold values
threshold_list = [i/10 for i in range(10)]
threshold_list
```

```python
# checking the pred_dataframe_new
pred_dataframe_new.head()
```

```python
# Using these different threshold values and making predictions of class
for i in threshold_list:
    pred_dataframe_new[i] = pred_dataframe_new['Prediction_Prob'].apply(lambda x: 1 if x>= i else 0)
pred_dataframe_new.head()
```

```python
# creating a new dataframe which includes cutoff, accuracy, sensitivity and specificity
model_dataframe = pd.DataFrame()
accuracy_list = []
sensitivity_list = []
specificity_list = []
for i in threshold_list:
    matrix = metrics.confusion_matrix(pred_dataframe_new['Actual'],pred_dataframe_new[i])
    TN = matrix[0,0]
    TP = matrix[1,1]
    FP = matrix[0,1]
    FN = matrix[1,0]
    accuracy_list.append((TN+TP)/(TN+TP+FP+FN))
    sensitivity_list.append(TP/(TP+FN))
    specificity_list.append(TN/(TN+FP))
model_dataframe['cutoff_prob'] = threshold_list
model_dataframe['accuracy'] = accuracy_list
model_dataframe['sensitivity'] = sensitivity_list
model_dataframe['specificity'] = specificity_list
model_dataframe
```

```python
# plotting the optimum estimation curve using the above dataframe
plt.figure()
plt.plot(model_dataframe['cutoff_prob'], model_dataframe['accuracy'])
plt.plot(model_dataframe['cutoff_prob'],model_dataframe['sensitivity'])
plt.plot(model_dataframe['cutoff_prob'],model_dataframe['specificity'])
plt.xlabel('cutoff probabilities')
plt.title('optimum value estimation')
plt.xlabel('accuracy/sensitivity/specificity')
plt.xticks(model_dataframe['cutoff_prob'])
plt.legend(loc =0)
plt.show()
```

From above graph we can see that the optimum value of threshold probability can be choosen as 0.3. As at this value accuracy,sensitivity and specificity are considerably high and decent values. Hence 0.3 can be considered as optimum value. Also the properties of best model is also satisfied i.e. sensitivity is higher thereby specificity is also higher which in turn makes FPR lower. Hence in order to be a good model it should have high TPR and low FPR.

So far we have looked into model assessing paramters like sensitivity, specificity and accuracy. Apart from these metrics there are other metrics as well called precision and recall which are very similar to sensitivity and specificity. Precision is nothing but positive prediction power which accounts for number of positives predicted correctly out of total number of positive predictions. Whereas Recall and sensitivity both are same and accounts for number of actual positives correctly predicted by the model out of actual number of positives. We can see that as sensitivity and specificity has tradeoff among them as the cutoff value increases, similarly precision and recall have tradeoff as the cutoff value changes. Let's calculate precision and recall

```python
# checking pred_dataframe_new
pred_dataframe_new.head()
```

```python
# lets create a new dataframe which includes cutoff, accuracy, precision, recall
model_dataframe_new = pd.DataFrame()
model_dataframe_new['cutoff'] = threshold_list
accuracy_list_new = []
precision_list = []
recall_list = []
for i in threshold_list:
    matrix = metrics.confusion_matrix(pred_dataframe_new['Actual'],pred_dataframe_new[i])
    TP = matrix[1,1]
    TN = matrix[0,0]
    FP = matrix[0,1]
    FN = matrix[1,0]
    accuracy_list_new.append((TP+TN)/(TP+TN+FP+FN))
    precision_list.append(TP/(TP+FP))
    recall_list.append(TP/(TP+FN))
model_dataframe_new['accuracy'] = accuracy_list_new
model_dataframe_new['precision'] = precision_list
model_dataframe_new['recall'] = recall_list
model_dataframe_new
```

```python
# plotting the above dataframe
plt.figure()
plt.plot(model_dataframe_new['cutoff'],model_dataframe['accuracy'])
plt.plot(model_dataframe_new['cutoff'],model_dataframe_new['precision'])
plt.plot(model_dataframe_new['cutoff'],model_dataframe_new['recall'])
plt.xlabel('cutoff prob')
plt.ylabel('probabilities')
plt.xticks(model_dataframe['cutoff_prob'])
plt.title('trade-off between precision and recall')
plt.legend(loc = 0)
plt.show()
```

From the precision and recall model assessment we get the optimum cutoff value to be 0.42. It is the cutoff point where precision and recall intersect with other.
